```
import pandas as pd

ruta_archivo = "/content/compras_semana.csv"
datos = pd.read_csv(ruta_archivo)
datos["total"] = datos["precio"] * datos["cantidad"]

#Suma columna total
suma_compra = datos['total'].sum()
print(f"Total de compra: {suma_compra}")

#Promedio
promedio_compra = datos['total'].mean()
print(f"Promedio de compra: {promedio_compra}")

#mediana compras
mediana_compra = datos['total'].median
print(f"mediana de compra: {mediana_compra}")

registro_conteo = datos['total'].value_counts()
print(registro_conteo)

datos.describe()
```
